```python
# 1. Import libraries

import pandas as pd
import os
import shutil
from zipfile import ZipFile
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

# 2. Import the provided salary data from Excel

data_file = "employee data.xlsx"
sheet_name = "Total"

try:
    df = pd.read_excel(data_file, sheet_name=sheet_name)
    print(f"Loaded {data_file} (sheet: {sheet_name})")
except FileNotFoundError:
    raise FileNotFoundError(f"File {data_file} not found. Please ensure it's in the working directory.")
except Exception as e:
    raise ValueError(f"Error loading Excel: {e}")

# Clean up column names (strip spaces, make consistent)
df.columns = df.columns.str.strip().str.replace(' ', '')

# Display first few rows
display(df.head())
```
